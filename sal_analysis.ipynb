```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the data is in a DataFrame called 'df'
df = pd.read_excel('Data_Scientist - Detailed.xlsx', sheet_name='Data_Scientist')  # Replace with actual data loading method
print(df.columns)
# Calculate average salary
df['Average Salary'] = (df['Salary Range (Min)'] + df['Salary Range (Max)']) / 2

# Group by experience and calculate mean salary
salary_by_experience = df.groupby('Experience')['Average Salary'].mean().reset_index()

# Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Experience', y='Average Salary', data=salary_by_experience)
sns.scatterplot(x='Experience', y='Average Salary', data=df, color='red', alpha=0.5)

plt.title('Average Salary Distribution by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary (SGD)')

# Add a trend line
sns.regplot(x='Experience', y='Average Salary', data=salary_by_experience, scatter=False, color='green')

plt.tight_layout()
plt.show()

# Print summary statistics
print(salary_by_experience)
print("\nOverall salary statistics:")
print(df['Average Salary'].describe())
```
